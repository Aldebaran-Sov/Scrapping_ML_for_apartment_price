```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('clean_cost_of_living.csv')
```

```python
df = df[df['Country'].isin(['United States', 'United Kingdom', 'Germany', 'France', 'Canada', 'Australia'])]
```

```python

for item in df["Item"].unique()[5:10]:
    item_data = df[df["Item"] == item]  # Filter data for the item
    plt.figure(figsize=(8, 5))
    plt.bar(item_data["Country"], item_data["Price"], color="skyblue")
    plt.title(f"Price of {item} by Country", fontsize=14)
    plt.xlabel("Country", fontsize=12)
    plt.ylabel("Price ($)", fontsize=12)
    plt.xticks(rotation=45, ha="right")
    plt.tight_layout()
    plt.show()
```
